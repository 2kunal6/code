```
import torch
import torchvision
from torchvision import datasets, transforms
```

```
# transform = ... converts data in MNIST to tensor (pytorch only works with tensors)
train = torchvision.datasets.MNIST("", train=True, transform=transforms.Compose([transforms.ToTensor()]), download=True)

test = torchvision.datasets.MNIST("", train=False, transform=transforms.Compose([transforms.ToTensor()]), download=True)
```

```
trainset = torch.utils.data.DataLoader(train, batch_size=10, shuffle=True)

testset = torch.utils.data.DataLoader(test, batch_size=10, shuffle=True)
```
